```python
import numpy as np
import random
import math 
import matplotlib.pyplot as plt 
```

```python
def fact(num):
    fact=1
    while num>1:
        fact*= num
        num-=1
    return fact


def main():
    noOfVal = 10000
    lambdaVal = 9.0
    vals= list()
    trials=1000
    initial=0.1
    increment = 0.01
    for i in range(noOfVal):
        vals.append(round(initial,3)) # appending values like 0.01,0.10,0.19 ... 100.01 lambda values basically
        initial+=increment
    
    random.shuffle(vals) # shuffling the lambda values
    r=9
    rFact = fact(r)
    points = dict()
    lnFact = math.log(rFact)
    
    for i in range(trials): # in each trial we choose a lambda value and compute the logarithm of the likelihood

        rando = random.randint(0,len(vals)-1)
        lambdaTemp = vals[rando]

        log = -lambdaTemp - lnFact + r * math.log(lambdaTemp)
        points[lambdaTemp] = log
        
    p1=[]
    p2=[]
    for i in sorted(points.keys()):
        p1.append(i)
        p2.append(points[i])
    #we plot the lambda values and likelihood value
    print(p1[p2.index(max(p2))],"lambda for max likelihood")
    plt.plot(p1,p2)
    plt.show()








    
main()
```

```python
def fact(num):
    fact=1
    while num>1:
        fact*= num
        num-=1
    return fact

def main():
    b=13
    noOfVal = 10000
    lambdaVal = 15.0
    vals= list()
    trials=1000
    initial=0.1
    increment = 0.01
    
    for i in range(noOfVal):
        vals.append(round(initial,3)) # appending values like 0.01,0.10,0.19 ... 9.01 lambda values basically
        initial+=increment
    
    random.shuffle(vals) # shuffling the lambda values
    r=9
    rFact = fact(r)
    points = dict()
    lnFact = math.log(rFact)
    for i in range(trials): # in each trial we choose a lambda value and compute the logarithm of the likelihood
        rando = random.randint(0,len(vals)-1)
        lambdaTemp = vals[rando]

        log = -(lambdaTemp+b) - lnFact + r * math.log(lambdaTemp+b)
        points[lambdaTemp] = log
        
    p1=[]
    p2=[]
    for i in sorted(points.keys()):
        p1.append(i)
        p2.append(points[i])
    #we plot the lambda values and likelihood values
    print(p1[p2.index(max(p2))],"lambda for max likelihood") # prints the value at which the likelihood is maximum
    plt.plot(p1,p2)
    plt.show()
main()

```
